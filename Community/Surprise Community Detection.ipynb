```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')

G = nx.from_pandas_edgelist(df, 'source', 'target')

communities = nx.community.label_propagation.label_propagation_communities(G)
communities_list = list(communities)

sorted_communities = sorted(communities_list, key=len, reverse=True)

for i in range(min(10, len(sorted_communities))):
    print(f"Community {i + 1}: {len(sorted_communities[i])} nodes")

pos = nx.spring_layout(G)
for i, community in enumerate(sorted_communities):
    nx.draw_networkx_nodes(G, pos, nodelist=community, node_size=50, label=f'Community {i + 1}')

nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```
