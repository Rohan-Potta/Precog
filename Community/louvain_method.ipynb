```python
import networkx as nx
import community  # python-louvain
import pandas as pd
import matplotlib.pyplot as plt

def create_graph_from_csv(file_path):
    df = pd.read_csv(file_path)
    G = nx.from_pandas_edgelist(df, 'source', 'target')
    return G

def louvain_method(graph):
    partition = community.best_partition(graph)
    return partition

def get_top_communities(communities, n=10):
    community_sizes = {}
    for node, community_id in communities.items():
        if community_id not in community_sizes:
            community_sizes[community_id] = 1
        else:
            community_sizes[community_id] += 1

    sorted_communities = sorted(community_sizes.items(), key=lambda x: x[1], reverse=True)
    top_communities = sorted_communities[:n]
    return top_communities

def visualize_community(graph, community_id):
    community_nodes = [node for node, c in communities.items() if c == community_id]
    subgraph = graph.subgraph(community_nodes)

    pos = nx.spring_layout(subgraph) 
    nx.draw(subgraph, pos, with_labels=True, node_color='skyblue', node_size=500, font_size=8, font_color='black', font_weight='bold')
    plt.title(f"Community {community_id} Visualization")
    plt.show()


graph = create_graph_from_csv('data.csv')
communities = louvain_method(graph)
top_communities = get_top_communities(communities, n=10)
for i, (community_id, size) in enumerate(top_communities, start=1):
    print(f"\nCommunity {i}: Size {size}, Members: {[(node, c) for node, c in communities.items() if c == community_id]}")
    visualize_community(graph, community_id)
```
