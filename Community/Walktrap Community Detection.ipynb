```python
import networkx as nx
import pandas as pd
import community 

df = pd.read_csv('data.csv')
G = nx.from_pandas_edgelist(df, 'source', 'target')
partition = community.best_partition(G, resolution=1.0)


nx.set_node_attributes(G, partition, 'community')
community_sizes = {i: len([node for node, comm in partition.items() if comm == i]) for i in set(partition.values())}
sorted_communities = sorted(community_sizes.items(), key=lambda x: x[1], reverse=True)

print("Top 10 communities:")
for i, (community_id, size) in enumerate(sorted_communities[:10]):
    print(f"Community {i + 1}: Size {size}")
print("\nNode community assignments:")
for node, community_id in partition.items():
    print(f"Node {node}: Community {community_id}")
```
