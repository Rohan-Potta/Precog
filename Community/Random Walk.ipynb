```python
import networkx as nx
import pandas as pd
import numpy as np

file_path = 'data.csv' 
df = pd.read_csv(file_path)


G = nx.from_pandas_edgelist(df, 'source', 'target')


def random_walk_community_detection(graph, num_walks=10, walk_length=30, seed=42):
    np.random.seed(seed)
    nodes = list(graph.nodes())
    node_communities = {}

    for i in range(num_walks):
        current_node = np.random.choice(nodes)
        for _ in range(walk_length):
            neighbors = list(graph.neighbors(current_node))
            if neighbors:
                current_node = np.random.choice(neighbors)
            else:
                break
            if current_node not in node_communities:
                node_communities[current_node] = i

    return node_communities

communities = random_walk_community_detection(G)

for node, community in communities.items():
    print(f'Node {node}: Community {community + 1}')

```
