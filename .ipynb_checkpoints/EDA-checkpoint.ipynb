```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('data_sample_10.csv', names=['source', 'target'])

# Create a directed graph using NetworkX
G = nx.from_pandas_edgelist(df, 'source', 'target', create_using=nx.DiGraph())

# Basic graph information
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())

# Calculate some graph metrics
print("Degree Centrality:", nx.degree_centrality(G))
print("Betweenness Centrality:", nx.betweenness_centrality(G))


```

```python
# Degree distribution
degrees = [G.degree(node) for node in G.nodes()]
plt.hist(degrees, bins=20, alpha=0.75)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```

```python
# Shortest path lengths
shortest_paths = nx.all_pairs_shortest_path_length(G)
all_lengths = [length for source, lengths in shortest_paths for target, length in lengths.items()]
plt.hist(all_lengths, bins=20, alpha=0.75)
plt.title("Shortest Path Lengths Distribution")
plt.xlabel("Shortest Path Length")
plt.ylabel("Frequency")
plt.show()
```

```python
# Clustering coefficient distribution
clustering_coefficients = nx.clustering(G)
plt.hist(list(clustering_coefficients.values()), bins=20, alpha=0.75)
plt.title("Clustering Coefficient Distribution")
plt.xlabel("Clustering Coefficient")
plt.ylabel("Frequency")
plt.show()
```

```python
# Strongly connected components
strongly_connected_components = nx.strongly_connected_components(G)
sizes = [len(component) for component in strongly_connected_components]
plt.hist(sizes, bins=20, alpha=0.75)
plt.title("Sizes of Strongly Connected Components")
plt.xlabel("Component Size")
plt.ylabel("Frequency")
plt.show()
```

```python
# Temporal degree distribution
degrees_over_time = [dict(G.degree(nodes)) for nodes in nx.connected_components(G.to_undirected())]
degree_df = pd.DataFrame(degrees_over_time)
degree_df.plot(kind='line', title='Temporal Degree Distribution')
plt.xlabel('Time')
plt.ylabel('Degree')
plt.show()
```

```python
connected_components = list(nx.strongly_connected_components(G))

# Calculate and plot temporal diameter for each connected component
for i, component in enumerate(connected_components):
    subgraph = G.subgraph(component)
    
    # Skip small connected components (optional)
    if len(subgraph) < 3:
        continue
    
    # Calculate temporal diameter
    diameters_over_time = [nx.diameter(subgraph.subgraph(nodes)) for nodes in nx.connected_components(subgraph.to_undirected())]
    
    # Plot the temporal diameter
    plt.plot(diameters_over_time, marker='o', label=f'Component {i + 1}')

plt.title('Temporal Diameter for Strongly Connected Components')
plt.xlabel('Time')
plt.ylabel('Diameter')
plt.legend()
plt.show()
```

```python
# Temporal connectivity
connectivity_over_time = [nx.node_connectivity(G.subgraph(nodes)) for nodes in nx.connected_components(G.to_undirected())]
plt.plot(timestamps, connectivity_over_time, marker='o')
plt.title('Temporal Connectivity')
plt.xlabel('Time')
plt.ylabel('Connectivity')
plt.show()
```
